# Lecture 08

## Exercise 08.1

In this lecture we want to variationally optimize the ground state of a single quantum particle in a one dimensional (1D) space confined by the external potential
$$
V(x) = x^4 - \frac{5}{2}x^2
$$
In order to obtain an approximate wave function for the ground state, the Variational Monte Carlo method uses a trial wave function, which for the ground state we choose to be
$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$

We write a C++ Variational Monte Carlo code for a single quantum particle in 1D which exploits the Metropolis algorithm to sample a trial wave function $|\Psi_T(x)|^2$ using an uniform transition probability $T(x_{new}|x_{old})$, starting from the already developed code used in lecture 05 for the hydrogen atom. Using data blocking, the code is able to compute the expectation value for the Hamiltonian.

## Exercise 08.2

The parameters $\sigma$ and $\mu$ in the wave function $\Psi_T^{\sigma,\mu}(x)$ need to minimize $\langle {\hat H} \rangle_T$ in order to approximate the ground state $E_0$:

$$
\langle {\hat H} \rangle_T = \frac{\langle \Psi_T^{\sigma,\mu}| {\hat H} | \Psi_T^{\sigma,\mu} \rangle}
{\langle \Psi_T^{\sigma,\mu} | \Psi_T^{\sigma,\mu} \rangle} \ge E_0 =
\frac{\langle \Psi_0| {\hat H} | \Psi_0 \rangle}
{\langle \Psi_0 | \Psi_0 \rangle}
$$

By cycling various possible values of these parameters, the best ones found are $\mu=0.805$ and $\sigma=0.610$.

We can now compute the value of $\langle {\hat H} \rangle_T$ for this choice of parameters running the algorithm for $100$ blocks of $10^4$ steps each and then confront this result with the numerical solution obtained by transforming the Schrödinger equation into a matrix equation like so:

```python
mu=0.805
sigma=0.610

def Vpot(x):
    return (x**2 - 2.5)*x**2

hbar = 1
m = 1
a = 6
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)
```

```python
block,H,H_err=np.loadtxt("data/energy.dat", usecols=(0,1,2), delimiter=" ", unpack='true')
print("Ground state energy: ", E[0])
print("VMC ground state energy: ", H[99], "±", H_err[99]) 

plt.figure(0,figsize=(15,7))
plt.hlines(E[0], 0, len(block), color='red', label='numerical ground state')
plt.errorbar(block, H, yerr=H_err, label='evaluated ground state')
plt.legend()
plt.xlabel("# of blocks")
plt.ylabel(r"$<H>_T$")
plt.grid(True)
plt.show()
```

We can observe that the evaluation of the ground state energy is not compatible with the numerical obtained one. This is to be expected as the variational method used is only able to give an approximation of such state.

In the next graph we plot the different probability distributions of the particle: the one obtained with the numerical method, the VMC sampled one and the analytic curve of the trial wave function using the chosen parameters.

```python
def psiT(x):
    t1=((x-mu)/sigma)**2
    t2=((x+mu)/sigma)**2
    return np.exp(-t1/2)+np.exp(-t2/2)

psiVMC=np.loadtxt("data/points.dat", usecols=(0), unpack='true')
plt.figure(figsize=(15,7))
bins=200
psi_opt=(psiT(x))**2
N=np.sum(dx*psi_opt)

plt.plot(x,(psi[0])**2, label='Numerical')
plt.hist(psiVMC,bins,range=(-3,3),label='VMC sampled',density='true')
plt.plot(x,psi_opt/N, label='VMC optimized')
plt.legend()
plt.xlabel("x")
plt.ylabel(r"$\rho(x)$")
plt.grid(True)
plt.show()
```

We can see that even though the probability distribution that we found with the variational method is not compatible with the one obtained with the numerical method, the distributions are still similar in the shape. The only way to get a better result with this technique would be to use a more appropriate trial wave function.

## Exercise 08.3

At this point, we can try to simulate the same system with the PIMC/PIGS code provided. This will be done using both a constant trial wave function $\Psi_T(x)=1$ and the one we used before $\Psi_T(x) = \Psi_T^{\sigma,\mu}(x)$ with the variational parameters we have chosen. We first compute the ground state ($T=0$) with the Path Integral Ground State code, using imaginary time $\tau=8.0$ and $300$ timeslices.

```python
timeslices,ekin_const,ekin_err_const=np.loadtxt("data/CONSTkinetic.dat", delimiter=" ", usecols=(0,1,2), dtype="int,float,float", unpack=True)
timeslices,epot_const,epot_err_const=np.loadtxt("data/CONSTpotential.dat", delimiter=" ", usecols=(0,1,2), dtype="int,float,float", unpack=True)
timeslices,ekin_trial,ekin_err_trial=np.loadtxt("data/TRIALkinetic.dat", delimiter=" ", usecols=(0,1,2), dtype="int,float,float", unpack=True)
timeslices,epot_trial,epot_err_trial=np.loadtxt("data/TRIALpotential.dat", delimiter=" ", usecols=(0,1,2), dtype="int,float,float", unpack=True)
etot_const = ekin_const + epot_const
etot_err_const = np.sqrt(ekin_err_const**2+epot_err_const**2)
etot_trial = ekin_trial + epot_trial
etot_err_trial = np.sqrt(ekin_err_trial**2+epot_err_trial**2)

plt.figure(figsize=(15,7))

plt.subplot(121)
plt.title("Constant wave function")
plt.errorbar(timeslices,ekin_const,yerr=ekin_err_const, label="Kinetic energy")
plt.errorbar(timeslices,epot_const,yerr=epot_err_const, label="Potential energy")
plt.errorbar(timeslices,etot_const,yerr=etot_err_const, label="Total energy",color="C3")
plt.axhline(E[0],0,1, label="Ground state energy",color="C2")
plt.grid()
plt.xlabel(r"Timeslice")
plt.ylabel("Energy")
plt.legend()

plt.subplot(122)
plt.title("Optimized wave function")
plt.errorbar(timeslices,ekin_trial,yerr=ekin_err_trial, label="Kinetic energy")
plt.errorbar(timeslices,epot_trial,yerr=epot_err_trial, label="Potential energy")
plt.errorbar(timeslices,etot_trial,yerr=etot_err_trial, label="Total energy",color="C3")
plt.axhline(E[0],0,1, label="Ground state energy",color="C2")
plt.grid()
plt.xlabel(r"Timeslice")
plt.ylabel("Energy")
plt.legend()

plt.show()
```

This time the total energy we obtain is clearly comparable with the one we got from the numerical method. In the next graph we show a confront between the probability distributions.

```python
x_const,psi_const,err_const=np.loadtxt("data/CONSTprobability.dat", usecols=(0,1,2), unpack='true')
x_trial,psi_trial,err_trial=np.loadtxt("data/TRIALprobability.dat", usecols=(0,1,2), unpack='true')
plt.figure(0,figsize=(15,7))

plt.plot(x,(psi[0])**2, label='Numerical')
plt.hist(psiVMC,bins,range=(-3,3),label='VMC sampled',density='true')
plt.errorbar(x_const,psi_const,yerr=err_const,label='PIGS constant')
plt.errorbar(x_trial,psi_trial,yerr=err_trial,label='PIGS optimized')
plt.legend()
plt.xlabel("x")
plt.ylabel(r"$\rho(x)$")
plt.xlim((-3,3))
plt.grid(True)
plt.show()
```

Again, the PIGS code is way better at finding the actual ground state. We can now try to change the value of $\tau$ and see what happens while using the constant trial wave function and while using the optimized one.

```python
x_c8, psi_c8, err_c8= np.loadtxt("data/CONSTprobabilityit8.dat", usecols=(0,1,2), unpack='true')
x_c2, psi_c2, err_c2= np.loadtxt("data/CONSTprobabilityit2.dat", usecols=(0,1,2), unpack='true')
x_c1, psi_c1, err_c1= np.loadtxt("data/CONSTprobabilityit1.dat", usecols=(0,1,2), unpack='true')
x_c01,psi_c01,err_c01=np.loadtxt("data/CONSTprobabilityit01.dat", usecols=(0,1,2), unpack='true')

plt.figure(figsize=(15,7))

plt.plot(x,(psi[0])**2, label='numerical')
plt.errorbar(x_c8,psi_c8,yerr=err_c8,label='$it=8$')
plt.errorbar(x_c2,psi_c2,yerr=err_c2,label='$it=2$')
plt.errorbar(x_c1,psi_c1,yerr=err_c1,label='$it=1$')
plt.errorbar(x_c01,psi_c01,yerr=err_c01,label='$it=0.1$')

plt.legend()
plt.xlabel("x")
plt.ylabel(r"$\rho(x)$")
plt.title(f"Constant wave function")
plt.xlim((-3,3))
plt.grid(True)
plt.show()

x_t8, psi_t8, err_t8= np.loadtxt("data/TRIALprobabilityit8.dat", usecols=(0,1,2), unpack='true')
x_t2, psi_t2, err_t2= np.loadtxt("data/TRIALprobabilityit2.dat", usecols=(0,1,2), unpack='true')
x_t1, psi_t1, err_t1= np.loadtxt("data/TRIALprobabilityit1.dat", usecols=(0,1,2), unpack='true')
x_t01,psi_t01,err_t01=np.loadtxt("data/TRIALprobabilityit01.dat", usecols=(0,1,2), unpack='true')

plt.figure(figsize=(15,7))

plt.plot(x,(psi[0])**2, label='numerical')
plt.errorbar(x_t8,psi_t8,yerr=err_t8,label='$it=8$')
plt.errorbar(x_t2,psi_t2,yerr=err_t2,label='$it=2$')
plt.errorbar(x_t1,psi_t1,yerr=err_t1,label='$it=1$')
plt.errorbar(x_t01,psi_t01,yerr=err_t01,label='$it=0.1$')

plt.legend()
plt.xlabel("x")
plt.ylabel(r"$\rho(x)$")
plt.title(f"Optimized wave function")
plt.xlim((-3,3))
plt.grid(True)
plt.show()
```

We can observe that the result obtained from the constant trial wave function quickly diverges from the actual ground state if we lower the imaginary time. Using the optimized trial function instead we can still have a good enough result for $\tau=1$. With $\tau=0.1$ the probability distribution starts to be significally different from the numerical one even with this trial wave function.


At last, we can see what happens using the Path Integral Monte Carlo code (with trial function $\Psi_T^{\sigma,\mu}(x)$) for temperatures different than zero (as the PIGS code only works for $T=0$):

```python
x_125,psi_125,err_125=np.loadtxt("data/TRIALprobabilityT1_25.dat", usecols=(0,1,2), unpack='true')
x_25,psi_25,err_25=np.loadtxt("data/TRIALprobabilityT2_5.dat", usecols=(0,1,2), unpack='true')
x_5,psi_5,err_5=np.loadtxt("data/TRIALprobabilityT5.dat", usecols=(0,1,2), unpack='true')
x_10,psi_10,err_10=np.loadtxt("data/TRIALprobabilityT10.dat", usecols=(0,1,2), unpack='true')

plt.figure(0,figsize=(15,7.5))

plt.plot(x,(psi[0])**2, label="Numerical")
plt.errorbar(x_125,psi_125,yerr=err_125,label='$T=1.25$')
plt.errorbar(x_25,psi_25,yerr=err_25,label='$T=2.5$')
plt.errorbar(x_5,psi_5,yerr=err_5,label='$T=5.0$')
plt.errorbar(x_10,psi_10,yerr=err_10,label='$T=10.0$')

plt.legend()
plt.xlabel("x")
plt.ylabel(r"$\rho(x)$")
plt.xlim((-3,3))
plt.grid(True)
plt.show()
```

Raising the temperature the probability distribution spreads more because the system has access to more excited states that contribute to the total wave function and this causes higher uncertainty on the position of the particle.
