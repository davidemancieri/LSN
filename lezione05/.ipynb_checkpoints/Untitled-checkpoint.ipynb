```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, sum_prog, err_prog = np.loadtxt( "mean100.out", usecols=(0,1,2), delimiter=' ', unpack='true' )

fig=plt.figure(figsize=(15,5))
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.title('Media')
plt.xlabel('#lanci')
plt.ylabel('$<r>$')
plt.grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

X,Y,Z=np.loadtxt( "points100.out", usecols=(0,1,2), delimiter=' ', unpack='true' )

fig = plt.figure(figsize=(10,8))
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, sum_prog = np.loadtxt( "EQvalues100.out", usecols=(0,1), delimiter=' ', unpack='true' )

fig=plt.figure(figsize=(15,5))
plt.plot(x,sum_prog)
plt.title('Media')
plt.xlabel('#lanci')
plt.ylabel('$<r>$')
plt.grid(True)
plt.show()
```

```python
R = np.loadtxt('EQvalues100.out', usecols=(1), max_rows=500)
autocorrelation = np.correlate(R,R, mode="full")
autocorrelation = autocorrelation[autocorrelation.size//2:]

plt.figure(figsize = (10,7))    
plt.plot(range(len(autocorrelation)),autocorrelation/1600)
plt.title('Autocorrelation Function for Radius',fontsize = 18)
plt.xlabel('Time step t',fontsize = 14)
plt.ylabel("Autocorrelation Ac(t)",fontsize = 14)
plt.grid()
plt.show()
```
