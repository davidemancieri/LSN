## Exercise 09.1

The aim of this lecture is to find a solution for the traveling salesman problem using a genetic algorithm. The individuals that are part of the population are possible paths that the salesman takes to visit all of the $N=32$ cities (each can be visited only once and the path must start and end on the same city). The fitness of each individual in the population is measured with the loss/cost function:
$$
L^{(1)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|
$$
Following the logic behind the genetic algorithms, the fittest individuals are chosen in pairs to crossover (with a probability of $65\%$. In case the crossover fails, the individuals chosen to be parents are reintroduced in the next generation) and generate an offspring of two new individuals that inherit characteristics from their parents and take their spot in the next generation. The parents are chosen by (after ordering every individual in the population by their fitness) taking the $j_{th}$ individual in the list, where $j=int(N\times r^p)$, with $r$ being a random number extracted uniformly in the interval $[0,1]$ and $p=2$. This ensures that the fittest individuals are chosen more often as $j$ will be more likely a small number compared to the number of individuals. An individual can be chosen multiple times as a parent while the new generation is being created. The newly generated individuals (or the chosen parents if no crossover occurred) are then "mutated", meaning that one or more of the following actions could be performed on the list of cities visited in that particular path:
1. pair permutation
2. shift of a block of cities
3. block permutation
4. inversion in the order of some cities

We chose the probability of each of these mutations to happen to be $9\%$ (independently).


Now that the code is ready, we can test if it is able to find the best path for an easy configuration of cities: we randomly generate them on a circle of radius $1$. We let the code run for $2000$ generations of $300$ individuals each.

```python
c_best,c_ave=np.loadtxt("circle_fitness.dat", usecols=(0,1), delimiter="   ", unpack='true')
c_x,c_y=np.loadtxt("circle_bestpath.dat", usecols=(0,1), delimiter="   ", unpack='true')
plt.figure(figsize=(14,7))
c_gen=np.linspace(1,2000,2000)

plt.subplot(121)
plt.plot(c_gen,c_best,label="Best path")
plt.plot(c_gen,c_ave,label="Average")
plt.title("Best path and average of the better half of paths of that generation")
plt.xlabel("Generation")
plt.ylabel("$L^{(1)}$")
plt.legend()
plt.grid(True)

plt.subplot(122)
plt.plot(c_x,c_y,'-o',label='Path')
plt.legend()
plt.title("Best path")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.show()

print("The best path found has length:", c_best[1999])
```

The algorithm clearly found the best path in this simple case (we also got a bit lucky as it managed to do so in about $100$ generations), which is obviously the one following the circumference. The fact that the average value of $L^{(1)}$ of the better half of the paths keeps oscillating shows that the algorithm is correctly changing the individuals at each generation, by crossover or mutation.
