### <span> a) Calcolo del valor medio </span>
Viene richiesto di computare una stima del valore di aspettazione di una variabile causale con distribuzione uniforme tra 0 e 1: $\langle r \rangle = \int_0^1 r dr$.

Per fare ciò, si generano M numeri casuali con tale distribuzione e ne si calcola dunque il valor medio come $\langle r \rangle = \frac{1}{M}\sum_{i=1}^{M} r_i$. Inoltre l'incertezza sul risultato è data dalla deviazione standard, pari a $$\sigma_r := \sqrt{\langle (r-\langle r\rangle)^2\rangle} = \sqrt{\langle r^2 \rangle -\langle r\rangle^2} $$
Si può quindi procedere come segue: l'insieme degli M numeri estratti precedentemente viene diviso in N blocchi da $L=\frac{M}{N}$ lanci ciascuno. In ognuno di questi blocchi viene calcolata una stima del valor medio di  $r$ e di $r^2$. Le medie su tutti gli M lanci sono quindi date da
$$\langle r\rangle^2 \simeq \left( \frac{1}{N} \sum_{j=1}^N r_j \right)^2 \quad \quad  \langle r^2 \rangle \simeq \frac{1}{N} \sum_{j=1}^N r^2_j $$
si vede quindi come l'incertezza statistica di $\langle r \rangle$ sia pari a
    $$\sigma \simeq \sqrt{ \frac{1}{N} \sum_{j=1}^N r^2_j - \left( \frac{1}{N} \sum_{j=1}^N r_j \right)^2} $$
È quindi ora possibile graficare la stima di $r$ all'aumentare dei blocchi presi in considerazione, assieme all'errore statistico ad essa associato.
Per ottenere i dati sono stati utilizzati i valori $M=10^6$ e $N=100$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, sum_prog, err_prog = np.loadtxt( "data_mean.out", usecols=(0,1,2), delimiter=' ', unpack='true' )

fig=plt.figure(figsize=(15,5))
plt.errorbar(x,sum_prog-0.5,yerr=err_prog)
plt.title('Media')
plt.xlabel('#lanci')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

Il grafico è in accordo con quanto ci si aspetta: $\langle r \rangle$ converge entro una deviazione standard a $\frac{1}{2}$, che è il risultato dell'integrale di semplice risoluzione analitica scritto in precedenza. Inoltre si osserva come l'errore tende a diminuire all'aumentare dei lanci.

### <span> b) Calcolo della varianza </span>
Viene richiesta ora una stima della varianza della distribuzione considerata in precedenza: $\sigma^2 = \int_0^1 (r-1/2)^2 dr$. Il metodo utilizzato è il medesimo del punto a.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, sum_prog, err_prog = np.loadtxt( "data_stdDev.out", usecols=(0,1,2), delimiter=' ', unpack='true' )

fig=plt.figure(figsize=(15,5))
plt.errorbar(x,sum_prog-1/12,yerr=err_prog)
plt.title('Varianza')
plt.xlabel('#lanci')
plt.ylabel('<(r-0.5)^2>-1/12')
plt.grid(True)
plt.show()
```

Anche in questo caso il grafico conferma quanto ci si aspettava. L'integrale può essere nuovamente calcolato analiticamente ottenendo come risultato $\frac{1}{12}$ e la varianza converge proprio a tale valore.

### <span> c) Test di Pearson </span>
È ora richiesto di utilizzare il test del $\chi^2$ per verificare che il generatore di numeri casuali utilizzato sia effettivamente uniforme.
La procedura è la seguente: si divide l'intervallo $[0,1]$ in $M$ sotto-intervalli identici e successivamente si generano $n$ numeri casuali. Si conta il numero $n_i$ di numeri che ricadono in ciascuno dei sotto-intervalli. Tale valore dovrebbe coincidere con il prodotto tra il numero di lanci totali $n$ e la probabilità che un numero ricada in quell'intervallo, che, supposto che la distribuzione sia effettivamente uniforme, è pari a $\frac{1}{M}$ per ciascuno di essi. Il $\chi^2$ è quindi dato da $$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}$$ e, poiché in media $(n_i - n/M)^2 \simeq n/M$, ci si aspetta che $\chi^2=M$.
Il test è stato ripetuto 100 volte, utilizzando $n=10^4$ e $M=100$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x,y = np.loadtxt( "data_chi2.out", usecols=(0,1), delimiter=' ', unpack='true' )

plt.errorbar(x,y)
plt.title('Test di Pearson')
plt.xlabel('#iterazioni')
plt.ylabel('chi2')
plt.grid(True)
plt.show()
```

I risultati ottenuti sono compatibili con quello atteso, pari a 100. Inoltre non si nota nessun andamento particolare, come è giusto che sia per una distribuzione davvero uniforme. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

x1=np.loadtxt("data_unif.out", skiprows=0, max_rows=10000, unpack=True)
x2=np.loadtxt("data_unif.out", skiprows=10000, max_rows=10000)
x10=np.loadtxt("data_unif.out", skiprows=20000, max_rows=10000)
x100=np.loadtxt("data_unif.out", skiprows=30000, max_rows=10000)

plt.subplot(221)
plt.hist(x1, 100, density=True)
plt.subplot(222)
plt.hist(x2, 100, density=True)
plt.subplot(223)
plt.hist(x10, 100, density=True)
plt.subplot(224)
plt.hist(x100, 100, density=True)

parameter = norm.fit(x100)
x = np.linspace(0.4,0.6,100)
y = norm.pdf(x, *parameter)
plt.plot(x,y)

plt.show()
```
