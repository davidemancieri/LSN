## Exercise 01.1
To test the correct behavior of the Pseudo-Random Number generator that we are using, we evaluate a simple integral of which we know the exact solution: $\langle r \rangle = \int_0^1 r dr = 1/2$.

We evaluate the integral $10^6$ times, in $100$ blocks of $10^4$ evaluations each. The blocking method is used to compute the uncertainty of the result. The following plot shows the estimated value of the integral as a function of the number of blocks considered, with the corresponding uncertainty.

```python
x,y,error=np.loadtxt("data_mean.out", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.figure(figsize=(10,5))
plt.hlines(0,0,len(x), color='red')
plt.errorbar(x,y-0.5,yerr=error)
plt.xlabel('# of blocks')
plt.ylabel(r"$\langle r \rangle - 1/2$")
plt.grid(True)

print("On the last block")
print("estimate:",y[99],"error:",error[99])

plt.show()
```

The estimated value of $\langle r \rangle$ was shifted down by $0.5$ for clarity. We can observe that the estimate is compatible with the exact result within the statistical error.

We now want to estimate the variance of the same probability distribution: $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$. To do so, we repeat what was done before, using again $100$ blocks of $10^4$ evaluations each. The following graph shows the result of such process.

```python
x,y,error=np.loadtxt("data_var.out", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.figure(figsize=(10,5))
plt.hlines(0,0,len(x), color='red')
plt.errorbar(x,y-1/12,yerr=error)
plt.xlabel('# of blocks')
plt.ylabel(r"$\sigma^2 - 1/12$")
plt.grid(True)

print("On the last block")
print("estimate:",y[99],"error:",error[99])

plt.show()
```

Again, the estimate of the variance was shifted down and converges to the expected value within statistical error.

Finally, we procede with the $\chi^2$ test on the generator, to reject the hypotesis that the generated numbers do not follow a uniform distribution. First, we divide the $[0,1]$ interval in $M$ sub-intervals. We then generate $n$ random numbers and count how many of those are in each sub-interval. If the generator is truly uniform, we should obtain that each sub-interval contains around $np = n\times 1/M= n/M$ numbers, where $p$ is the probability for a number to be in such sub-interval.

We can now compute the $\chi^2$ comparing the expected value $n/M$ with the observed ones $n_i$: $$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}$$
We use $M=100$ and $n=10^4$. We should expect on average $(n_i - n/M)^2 \simeq n/M$, which implies $\chi^2 \simeq 100$, i.e. the number of sub-intervals. A larger value of $\chi^2$ indicates that the hypothesis ($n_i$ are drawn from a uniform distribution) is rather unlikely.

We repeat the test $100$ times, using for the first test the first $10^4$ generated numbers, then the successive $10^4$ ones and so on.

```python
x,y=np.loadtxt("data_chi2.out", usecols=(0,1), delimiter=" ", unpack='true')
plt.figure(figsize=(10,5))
plt.hlines(100,0,len(x), color='red')
plt.plot(x,y, "o")
plt.xlabel('# of iteration')
plt.ylabel('$\chi^2$')
plt.grid(True)

plt.show()
```

As expected, the value of $\chi^2$ is ditribuited around the number of sub-intervals, $100$.
## Exercise 01.2
After using the method of the inversion of the cumulative function to extend the random number generator with the option to generate a number from a generic exponential distribution ( $p(x) = \lambda \exp(-\lambda x), x\in [0,+\infty]$ ) and a generic Cauchy-Lorentz one ( $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}, x\in [-\infty,+\infty]$ ), we want to check the validity of the Central Limit Theorem.

To do so, we fill histograms with $10^4$ realizations of $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ (for $N=1, 2, 10, 100$), $x_i$ being a random variable sampled throwing a standard (continous) dice, an exponential dice (we use $\lambda=1$) or a Lorentzian dice (we use $\mu=0$ and $\Gamma=1$).

```python
from scipy.stats import norm

x1=np.loadtxt("data_unif.out", skiprows=0, max_rows=10000, unpack=True)
x2=np.loadtxt("data_unif.out", skiprows=10000, max_rows=10000)
x10=np.loadtxt("data_unif.out", skiprows=20000, max_rows=10000)
x100=np.loadtxt("data_unif.out", skiprows=30000, max_rows=10000)

fig=plt.figure(figsize=(15,10))
fig.suptitle('Standard dice', fontsize=24)

plt.subplot(221)
plt.hist(x1, 100, density=True)
plt.title('N = 1')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.subplot(222)
plt.hist(x2, 100, density=True)
plt.title('N = 2')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.subplot(223)
plt.hist(x10, 100, density=True)
plt.title('N = 10')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.subplot(224)
plt.hist(x100, 100, density=True)
plt.title('N = 100')
plt.xlabel('x')
plt.ylabel('p(x)')

parameter = norm.fit(x100)
x = np.linspace(3,4,100)
y = norm.pdf(x, *parameter)
plt.plot(x,y)

plt.show()

print('Optimized parameters:')
print('Mu: ', parameter[0], 'Sigma:', parameter[1])
```

The Central Limit Theorem can be easily seen at work in the histograms above. It is in fact clear how the shape of the distribution of the average of the uniformly generated numbers gets more similar to a gaussian distribution as $N$ grows. In the case of $N=100$ the data was fitted with a gaussian profile to better show this. The CLT says that the gaussian limit distribution will have the same expectation value as the original distribution and variance $\sigma_{limit}^2=\frac{\sigma_{original}^2}{N}$. The expectation value of the original distribution in this case is
$$\mu=\frac{1}{5}\int_1^6x\,dx=\frac{7}{2}=3.5$$
and the variance
$$\sigma^2=\frac{1}{5}\int_1^6(x-\mu)^2\,dx=\frac{25}{12}$$
so 
$$\sigma_{limit}=\frac{\sigma_{original}}{\sqrt{N}}=\sqrt{\frac{25}{12\cdot100}}=\frac{\sqrt{3}}{12}\approx0.144$$
This means that the parameters obtained from the fitting are compatible with the expected ones ($\mu=3.5$, $\sigma\approx0.144$).

```python
x1=np.loadtxt("data_exp.out", skiprows=0, max_rows=10000, unpack=True)
x2=np.loadtxt("data_exp.out", skiprows=10000, max_rows=10000)
x10=np.loadtxt("data_exp.out", skiprows=20000, max_rows=10000)
x100=np.loadtxt("data_exp.out", skiprows=30000, max_rows=10000)

fig=plt.figure(figsize=(15,10))
fig.suptitle('Exponential dice', fontsize=24)

plt.subplot(221)
plt.hist(x1, 100, density=True)
plt.title('N = 1')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.subplot(222)
plt.hist(x2, 100, density=True)
plt.title('N = 2')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.subplot(223)
plt.hist(x10, 100, density=True)
plt.title('N = 10')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.subplot(224)
plt.hist(x100, 100, density=True)
plt.title('N = 100')
plt.xlabel('x')
plt.ylabel('p(x)')

parameter = norm.fit(x100)
x = np.linspace(0.6,1.4,100)
y = norm.pdf(x, *parameter)
plt.plot(x,y)

plt.show()

print('Optimized parameters:')
print('Mu: ', parameter[0], 'Sigma:', parameter[1])
```

The values extracted from the exponential distribution also confirm the CLT.

Again for $N=100$ a gaussian fit is shown. The expectation value of the original distribution this time is $\mu=\lambda^{-1}=1$ and the variance $\sigma^2=\lambda^{-2}=1$, so
$$\sigma_{limit}=\frac{\sigma_{original}}{\sqrt{N}}=\sqrt{\frac{1}{100}}=0.1$$
Once again this is compatible to what was obtained fitting the histrogram.

```python
from scipy.stats import cauchy

x1=np.loadtxt("data_lorentz.out", skiprows=0, max_rows=10000, unpack=True)
x2=np.loadtxt("data_lorentz.out", skiprows=10000, max_rows=10000)
x10=np.loadtxt("data_lorentz.out", skiprows=20000, max_rows=10000)
x100=np.loadtxt("data_lorentz.out", skiprows=30000, max_rows=10000)

fig=plt.figure(figsize=(15,10))
fig.suptitle('Lorentzian dice', fontsize=24)

plt.subplot(221)
plt.hist(x1, 100, range=(-10,10), density=True)
plt.title('N = 1')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.subplot(222)
plt.hist(x2, 100, range=(-10,10), density=True)
plt.title('N = 2')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.subplot(223)
plt.hist(x10, 100, range=(-10,10), density=True)
plt.title('N = 10')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.subplot(224)
plt.hist(x100, 100, range=(-10,10), density=True)
plt.title('N = 100')
plt.xlabel('x')
plt.ylabel('p(x)')

parameter = cauchy.fit(x100)
x = np.linspace(-10,10,100)
y = cauchy.pdf(x, *parameter)
plt.plot(x,y)

plt.show()

print('Optimized parameters:')
print('Mu: ', parameter[0], 'Gamma:', parameter[1])
```

The Cauchy-Lorentz distribution does not satisfy the hypotesis of the CLT (because its second moment is undefined) and so the distribution of the averaged values does not converge to a gaussian curve. The Cauchy-Lorentz distribution is in fact its own attractor, so the shape of the histogram is unchanged while increasing the number of averaged values.

For $N=100$ this time the data was fitted with a Cauchy-Lorentz profile. The optimized parameters are close to the original ones ($\mu=0, \Gamma=1$), proving again that the distribution does not change.
## Exercise 01.3
Now we want to simulate Buffon's experiment in order to obtain an estimated value for $\pi$. For the length of the needle we use $L=5$ and for the distance between lines $d=10$. The experiment is repeated $10^5$ times and the results divided into $100$ blocks of $10^3$ evaluations each.

```python
x,y,error=np.loadtxt("data_buffon.out", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.figure(figsize=(10,5))
plt.hlines(np.pi,0,len(x),color='red',label="$\pi$")
plt.errorbar(x,y,yerr=error)
plt.xlabel('# of blocks')
plt.ylabel('Estimate of $\pi$')
plt.grid(True)
plt.legend()

print("On the last block")
print("estimate:",y[99],"error:",error[99])

plt.show()
```

The estimated value converges within statistical uncertainty to the expected value.

In order to simulate the experiment, there is the need to randomly generate the position of the center of the needle between the interval $[0,d]$ and its inclination between the interval $[0,\pi]$. Since using $\pi$ to compute an estimate of $\pi$ itself would defeat the purpose of the experiment, the angle was not generated extracting it directly from $[0,\pi]$. Instead a couple of values $(x,y)$ were uniformly extracted ($x$ from the interval $[-1,1]$, $y$ from $[0,1]$). If $r^2=x^2+y^2\geq1$ a new couple was generated and this was repeated until $r^2<1$. The random point $(x,y)$ obtained is on an opened semidisk and can be described in the form $(r\cos\theta, r\sin\theta)$. We can thus get a value of $\theta$ uniformly distribuited between the interval $[0,\pi]$ from $\theta=\arccos(x/r)$. Using this technique we toss away around $1-\frac{\frac{\pi}{2}}{2}\approx 21.5\%$ points while looking for a good one.
