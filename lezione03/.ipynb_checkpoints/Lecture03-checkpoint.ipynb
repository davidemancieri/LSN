We want to compute at time $t=0$ via Monte Carlo the European call-option price, $C[S(0),0]$, and put-option price, $P[S(0),0]$ using the following parameters: 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

in two different ways:
1. by sampling directly the final asset price $S(T)$</span> for a geometric Brownian motion $GBM(r,\sigma^2)$
2. by sampling the discretized $GBM(r,\sigma^2)$ path of the asset price</span> 
dividing $[0,T]$ in $10^2$ time intervals: $S(0)$, $S(t_1)$, $S(t_2)$, ... up to $S(t_{100}=T=1)$

We make $10^5$ simulations in $100$ blocks of $10^3$ evaluations each. To check the results, we can compare with the analytic solution of the problem that follows.

```python
import scipy
from scipy.special import erf
from scipy import log, exp, sqrt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
```

```python
import numpy as np
import matplotlib.pyplot as plt

xC_direct,yC_direct, errorC_direct=np.loadtxt("data_callDirect.out", usecols=(0,1,2), delimiter=" ", unpack='true')
xC_discrt,yC_discrt, errorC_discrt=np.loadtxt("data_callDiscretized.out", usecols=(0,1,2), delimiter=" ", unpack='true')
xP_direct,yP_direct, errorP_direct=np.loadtxt("data_putDirect.out", usecols=(0,1,2), delimiter=" ", unpack='true')
xP_discrt,yP_discrt, errorP_discrt=np.loadtxt("data_putDiscretized.out", usecols=(0,1,2), delimiter=" ", unpack='true')

fig = plt.figure(figsize=(16,16))

plt.subplot(221)
plt.hlines(call,0,len(xC_direct), color='red',label="Analytic solution")
plt.errorbar(xC_direct,yC_direct,yerr=errorC_direct)
plt.title("Direct call-option price")
plt.xlabel('# of blocks')
plt.ylabel("C [S(0),0]")
plt.legend()
plt.grid(True)

plt.subplot(222)
plt.hlines(call,0,len(xC_discrt), color='red',label="Analytic solution")
plt.errorbar(xC_discrt,yC_discrt,yerr=errorC_discrt)
plt.title('Discrete call-option price')
plt.xlabel('# of blocks')
plt.ylabel("C [S(0),0]")
plt.legend()
plt.grid(True)

plt.subplot(223)
plt.hlines(put,0,len(xP_direct), color='red',label="Analytic solution")
plt.errorbar(xP_direct,yP_direct,yerr=errorP_direct)
plt.title('Direct put-option price')
plt.xlabel('# of blocks')
plt.ylabel("P [S(0),0]")
plt.legend()
plt.grid(True)

plt.subplot(224)
plt.hlines(put,0,len(xP_discrt), color='red',label="Analytic solution")
plt.errorbar(xP_discrt,yP_discrt,yerr=errorP_discrt)
plt.title('Discrete put-option price')
plt.xlabel('# of blocks')
plt.ylabel("P [S(0),0]")
plt.legend()
plt.grid(True)

plt.show()

print('On the last block')
print('Direct call-option price: ', yC_direct[99], '±', errorC_direct[99])
print('Discrete call-option price: ', yC_discrt[99], '±', errorC_discrt[99])
print('Direct put-option price: ', yP_direct[99], '±', errorP_direct[99])
print('Discrete put-option price: ', yP_discrt[99], '±', errorP_discrt[99])
print(" == BLACK-SCHOLES ==")
print("call: ", call)
print("put: ", put)
```

We can observe that the estimates are compatible with the analytic solutions within the statistical error. Both methods provide the result with approximately the same error, so the direct one should be preferred, as it is less computational expensive than the discrete one.
