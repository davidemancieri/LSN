## Exercise 02.1

We want to evaluate the following 1D integral via Monte Carlo 
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$
in two different ways:
1. sampling a uniform distribution in $[0,1]$
2. using importance sampling (i.e. sampling a non-uniform probability in $[0,1]$)

In both cases we evaluate the integral $10^6$ times and use the blocking method to compute the uncertainty of the result, with $100$ blocks of $10^4$ evaluations each. The following graph shows the result for uniform sampling.

```python
x1,y1,error1=np.loadtxt("data_unif.out", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.figure(figsize=(10,5))
plt.hlines(0,0,len(x1), color='red')
plt.errorbar(x1,y1-1,yerr=error1)
plt.title('Uniform sampling')
plt.xlabel('# of blocks')
plt.ylabel('I - 1')
plt.grid(True)

print("On the last block")
print("estimate:",y1[99],"error:",error1[99])

plt.show()
```

The estimated value of $I$ was shifted down by $1$ for clarity. We can observe that the estimate is compatible with the exact result within the statistical error.

To evaluate the integral using importance sampling instead, we need a function $p(x)$ similar to the one we want to integrate, such that on the integration interval $[0,1]$ $p(x)$ is positive and normalized ( $p(x)>0$ and $\int_0^1 p(x) dx = 1$ ).
We can use $p(x)=2(1-x)$ to keep things easy. The following graph shows the similarity between the functions.

```python
x = np.linspace(0,1,100)
I = np.pi/2. * np.cos(np.pi/2. * x)
S = 2 * (1-x)

plt.figure(figsize=(10,5))
plt.plot(x,I,label="Integrand")
plt.plot(x,S,label="p(x)")
plt.grid()
plt.legend()

plt.show()
```

We can use the method of the inversion of the cumulative distribution to find a way to generate random values following the distribution $p(x)$. First, we find the cumulative distribution:
$$F_x(x)=\int_0^x 2(1-x)dx=2x-x^2=:y$$
Some easy steps follow:
$$-(x^2-2x)=y$$
$$-(x^2-2x+1)+1=y$$
$$-(x-1)^2+1=y$$
$$x=1+\sqrt{1-y}$$
If $y$ is uniformly extracted in $[0,1]$, $x$ will be a random variable following the distribution $p(x)$.
At this point we can apply the importance sampling method to evaluate the integral.

```python
x2,y2,error2=np.loadtxt("data_samp.out", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.figure(figsize=(10,5))
plt.hlines(0,0,len(x2), color='red')
plt.errorbar(x2,y2-1,yerr=error2)
plt.title('Importance sampling')
plt.xlabel('# of blocks')
plt.ylabel('I - 1')
plt.grid(True)

print("On the last block")
print("estimate:",y2[99],"error:",error2[99])

plt.show()
```

Again, the estimated value of $I$ was shifted down by $1$ for clarity and we can observe that the estimate is compatible with the exact result within the statistical error.

It's interesting to observe that the error on the last block using the importance sampling method is about $4$ times smaller than the one using uniform sampling ($0.000134$ instead of $0.000501$). The relative error goes down from $0.050\%$ to $0.013\%$. We can visualize this better by plotting both methods in the same graph:

```python
plt.figure(figsize=(10,5))
plt.hlines(0,0,len(x1), color='red')
plt.errorbar(x1,y1-1,yerr=error1,label='Uniform sampling')
plt.errorbar(x2,y2-1,yerr=error2,label='Importance sampling')
plt.title('Comparison')
plt.xlabel('# of blocks')
plt.ylabel('I - 1')
plt.legend()
plt.grid(True)

plt.show()
```

## Exercise 02.2
We want to simulate a random walk (RW) in 3D on a cubic lattice and in the continuum. In both cases the RW is made of $100$ steps of length $1$. The RWs are repeated $10^4$ times each and the blocking method is applied, with $100$ blocks of $100$ simulations each.

The discrete RW can be realized simply by rolling a 6 face dice to decide in which direction to move (three axes, both in positive and negative direction). In the continuous case, one might think that the direction of the step can be chosen by extracting the two angles $\theta$ in $[0, \pi]$ and $\phi$ in $[0, 2\pi]$ and using the polar coordinates to express the direction, but this would generate a non uniform distribution on the sphere of radius $1$. The extracted points would in fact be crowded on the two poles because of the properties of the polar coordinates.

One way to actually generate points on a sphere uniformly is the following: we uniformly generate a point inside a cube $\vec{x} \in [-1,1]\times[-1,1]\times[-1,1]$ until we get one such that $|\vec{x}|< 1$. Then we normalize the vector to $1$ and this will give us the direction of the step on length $1$ that we wanted (correctly uniformly generated). With this method we reject $1-\frac{\frac{4}{3}\pi}{8}\approx 47.6\%$ points while looking for one we can use.

The following graphs show the distance from the origin as a function of the number of steps done during the RW. The distance is taken as the average of all $100$ blocks.

```python
xD,yD, errorD=np.loadtxt("data_discreteRW.out", usecols=(0,1,2), delimiter=" ", unpack='true')
xC,yC, errorC=np.loadtxt("data_continuumRW.out", usecols=(0,1,2), delimiter=" ", unpack='true')
fig = plt.figure(figsize=(16,8))

plt.subplot(121)
plt.errorbar(xD,yD,yerr=errorD)
plt.title('Discrete RW')
plt.xlabel('# of steps')
plt.ylabel(r'$\sqrt{\langle |\vec{r}|^2 \rangle}$')
plt.grid(True)

plt.subplot(122)
plt.errorbar(xC,yC,yerr=errorC)
plt.title('Continuous RW')
plt.xlabel('# of steps')
plt.ylabel(r'$\sqrt{\langle |\vec{r}|^2 \rangle}$')
plt.grid(True)

plt.show()

print("On the last step of the discrete RW")
print("estimate:",yD[99],"error:",errorD[99])
print("On the last step of the continuous RW")
print("estimate:",yC[99],"error:",errorC[99])
```

We can clearly see that both plots show a function like $f(N)=k\sqrt{N}$, so we can try to fit the data to find the best value for $k$.

```python
from scipy.optimize import curve_fit

def f(x,k):
    return k*np.sqrt(x)

p_optD,p_covD=curve_fit(f,xD,yD)
print("Optimized parameter k for the discrete RW = ", p_optD, '±' , np.sqrt(np.diagonal(p_covD)))

p_optC,p_covC=curve_fit(f,xC,yC)
print("Optimized parameter k for the continuous RW = ", p_optC, '±' , np.sqrt(np.diagonal(p_covC)))
```

Since the relative error on the distance is really small, a better way to visualize that $k$ is a good parameter fit is to plot $\sqrt{\langle |\vec{r}|^2 \rangle_{RW}}-k\sqrt{step}$.

```python
fig = plt.figure(figsize=(16,8))

plt.subplot(121)
plt.hlines(0,0,len(xD)-1, color='red')
plt.errorbar(xD,yD-(p_optD*np.sqrt(xD)),yerr=errorD)
plt.title('Discrete RW')
plt.xlabel('# of steps')
plt.ylabel(r'$\sqrt{\langle |\vec{r}|^2 \rangle}-k_D\sqrt{step}$')
plt.grid(True)

plt.subplot(122)
plt.hlines(0,0,len(xC)-1, color='red')
plt.errorbar(xC,yC-(p_optC*np.sqrt(xC)),yerr=errorC)
plt.title('Continuous RW')
plt.xlabel('# of steps')
plt.ylabel(r'$\sqrt{\langle |\vec{r}|^2 \rangle}-k_C\sqrt{step}$')
plt.grid(True)
```

It is now clear how the RW is compatible with the fitted function at every step. We can also notice that the error on the distance from the origin grows with the number of steps taken, this is rather obvious as the number of different possible positions goes up the more steps the walker takes.

The fact that the parameter of the fit is $1$ indicate diffusive behavior. In fact it can be shown that $\langle |\vec{r}(t)|^2 \rangle=2Dt$, where $D=\frac{a^2}{2\tau}$ is the diffusive parameter, with $a$ being the length of a step, $\tau$ the time between two consecutive steps and $t$ the total time. Since $t=N\tau$, 
$$\langle |\vec{r}(t)|^2 \rangle=2Dt=2\frac{a^2}{2\tau}N\tau=a^2N$$ so 
$$\sqrt{\langle |\vec{r}(t)|^2 \rangle}=a\sqrt{N}$$
In our case we have chosen $a=1$, so the fact that the fit parameter $k$ is close to $1$ confirms the diffusive behavior.
