## Exercise 07.1

The aim of the lecture is to implement the provided Monte Carlo NVT code to simulate a system of particles ($108$ particles of Argon, in the same phases as those described in the lecture 04) interacting following a Lennard-Jones potential in the canonical ensemble. In particular, we are interested in measuring the total energy and the pressure of the system.

The first thing to do is to check how many steps are required to equilibrate the system. To do that, we plot instantaneous values of the energy and pressure. The solid state is simulated starting from a fcc configuration, while the liquid one starts from an equilibrated solid state and the gas one from an equilibrated liquid state.

```python
ySE = np.loadtxt( "data/solidEQoutput.epot.0")
ySP = np.loadtxt( "data/solidEQoutput.pres.0")
yLE = np.loadtxt( "data/liquidEQoutput.epot.0")
yLP = np.loadtxt( "data/liquidEQoutput.pres.0")
yGE = np.loadtxt( "data/gasEQoutput.epot.0")
yGP = np.loadtxt( "data/gasEQoutput.pres.0")
x = np.linspace(1,len(ySE),num=len(ySE))

fig=plt.figure(figsize=(15,20))

plt.subplot(321)
plt.plot(x,ySE)
plt.title('Solid equilibration, energy')
plt.xlabel('# of step')
plt.ylabel(r'U/N [$\varepsilon$]')
plt.grid(True)

plt.subplot(322)
plt.plot(x,ySP)
plt.title('Solid equilibration, pressure')
plt.xlabel('# of step')
plt.ylabel(r'P [$\varepsilon/\sigma^3$]')
plt.grid(True)

plt.subplot(323)
plt.plot(x,yLE)
plt.title('Liquid equilibration, energy')
plt.xlabel('# of step')
plt.ylabel(r'U/N [$\varepsilon$]')
plt.grid(True)

plt.subplot(324)
plt.plot(x,yLP)
plt.title('Liquid equilibration, pressure')
plt.xlabel('# of step')
plt.ylabel(r'P [$\varepsilon/\sigma^3$]')
plt.grid(True)

plt.subplot(325)
plt.plot(x,yGE)
plt.title('Gas equilibration, energy')
plt.xlabel('# of step')
plt.ylabel(r'U/N [$\varepsilon$]')
plt.grid(True)

plt.subplot(326)
plt.plot(x,yGP)
plt.title('Gas equilibration, pressure')
plt.xlabel('# of step')
plt.ylabel(r'P [$\varepsilon/\sigma^3$]')
plt.grid(True)

plt.show()
```

We can observe that the equilibrium is reached quite fast, after about $250$ steps in the worst scenario. Because of this, we will use a $10^3$ steps equilibration routine at the beginning of every next simulation in this lecture.

Next, we want to know how big the blocks of our simulation should be. The first test that can be done is checking how quickly the autocorrelation function drops while enlarging the size of the blocks. We use $2\times10^5$ instantaneous values of pressure and energy for each state of the system.

```python
def AutoCorrelation(x, tau):
    
    l = len(x)-tau
    x_t = 0
    x_tau = 0
    x_ttau = 0
    x_t2 = 0
    
    for t in range(l):
        x_t  += x[t]
        x_tau += x[t+tau]
        x_ttau += x[t]*x[t+tau]
        x_t2  += x[t]**2
        
    x_t /= l
    x_tau /= l
    x_ttau /= l
    x_t2 /= l
    
    return (x_ttau-x_t*x_tau)/(x_t2-x_t**2)
```

```python
Usolid=np.loadtxt("data/solidautooutput.epot.0")
Psolid=np.loadtxt("data/solidautooutput.pres.0")
Uliquid=np.loadtxt("data/liquidautooutput.epot.0")
Pliquid=np.loadtxt("data/liquidautooutput.pres.0")
Ugas=np.loadtxt("data/gasautooutput.epot.0")
Pgas=np.loadtxt("data/gasautooutput.pres.0")

CorrUsolid = []
CorrPsolid = []
CorrUliquid = []
CorrPliquid = []
CorrUgas = []
CorrPgas = []
```

```python
print("This will take a while...")
for tau in range(100):
    
    CorrUsolid.append(AutoCorrelation(Usolid,tau))          #Internal energy autocorrelation
    CorrUliquid.append(AutoCorrelation(Uliquid,tau)) 
    CorrUgas.append(AutoCorrelation(Ugas,tau))
    
    CorrPsolid.append(AutoCorrelation(Psolid,tau))          #Pressure autocorrelation
    CorrPliquid.append(AutoCorrelation(Pliquid,tau))
    CorrPgas.append(AutoCorrelation(Pgas,tau))

tau = np.arange(100)
print("Done.")
```

```python
plt.figure(figsize = (15, 20))

plt.subplot(311)
plt.title("Autocorrelation solid phase")
plt.xlabel(r'$\tau$')
plt.ylabel('$Autocorrelation$')
plt.plot(tau,CorrUsolid,label="$U/N$")
plt.plot(tau,CorrPsolid,label="$Pressure$")
plt.grid(True)
plt.legend()

plt.subplot(312)
plt.title("Autocorrelation liquid phase")
plt.xlabel(r'$\tau$')
plt.ylabel('$Autocorrelation$')
plt.plot(tau,CorrUliquid,label="$U/N$")
plt.plot(tau,CorrPliquid,label="$Pressure$")
plt.grid(True)
plt.legend()

plt.subplot(313)
plt.title("Autocorrelation gas phase")
plt.xlabel(r'$\tau$')
plt.ylabel('$Autocorrelation$')
plt.plot(tau,CorrUgas,label="$U/N$")
plt.plot(tau,CorrPgas,label="$Pressure$")
plt.grid(True)
plt.legend()

plt.show()
```

The autocorrelation seems to drop quite fast, especially in the gas phase. However, given the long time needed to process the amount of data we have, it is hard to check for $\tau$s larger than $100$ and for the liquid phase the autocorrelation is still pretty high at that value, so we should try another technique.

In the next graphs, using the same set of data as above, we plot the statistical uncertainty on energy and pressure as a function of the size of the blocks for the three phases.

```python
def Error(x, L):
    
    M = len(x)
    N = int(M/L)
    
    ave = np.zeros(N)
    
    for i in range(N):
        
        blk_avg=0

        for j in range(L):
            blk_avg += x[j+L*i]
            
        ave[i] = blk_avg/L
        
    accu = 0
    accu2 = 0

    for i in range(N):
        accu += ave[i]
        accu2 += (ave[i])**2
    
    accu /= N
    accu2 /= N
        
    return np.sqrt(abs(accu2-(accu**2))/N)
```

```python
print("This will take about 1 minute...")

errSU = []
errSp = []
errLU = []
errLp = []
errGU = []
errGp = []

for blk_len in range(10, 5000, 50):
    errSU.append(Error(Usolid, blk_len))
    errSp.append(Error(Psolid, blk_len))
    errLU.append(Error(Uliquid, blk_len))
    errLp.append(Error(Pliquid, blk_len))
    errGU.append(Error(Ugas, blk_len))
    errGp.append(Error(Pgas, blk_len))

print("Done.")
```

```python
fig=plt.figure(figsize=(15,20))

plt.subplot(321)
plt.plot(range(10, 5000, 50), errSU)
plt.title('Solid energy uncertainty')
plt.xlabel('Block size')
plt.ylabel('Statistical uncertainty')
plt.grid(True)

plt.subplot(322)
plt.plot(range(10, 5000, 50), errSp)
plt.title('Solid pressure uncertainty')
plt.xlabel('Block size')
plt.ylabel('Statistical uncertainty')
plt.grid(True)

plt.subplot(323)
plt.plot(range(10, 5000, 50), errLU)
plt.title('Liquid energy uncertainty')
plt.xlabel('Block size')
plt.ylabel('Statistical uncertainty')
plt.grid(True)

plt.subplot(324)
plt.plot(range(10, 5000, 50), errLp)
plt.title('Liquid pressure uncertainty')
plt.xlabel('Block size')
plt.ylabel('Statistical uncertainty')
plt.grid(True)

plt.subplot(325)
plt.plot(range(10, 5000, 50), errGU)
plt.title('Gas energy uncertainty')
plt.xlabel('Block size')
plt.ylabel('Statistical uncertainty')
plt.grid(True)

plt.subplot(326)
plt.plot(range(10, 5000, 50), errGp)
plt.title('Gas pressure uncertainty')
plt.xlabel('Block size')
plt.ylabel('Statistical uncertainty')
plt.grid(True)

plt.show()
```

We can observe that in all the phases the statistical uncertainty at first grows really fast, reaching then a point where it oscillates around a value. Comparing these results with the plots of the autocorrelations above, we can clearly see that the block size at which the statistical uncertainty settles must coincide with the one at which the autocorrelation reaches $0$. Since this method takes less time than the computing of the autocorrelation function, we are able to check for bigger block sizes and see that the liquid phase requires blocks of at least about $400$ length.

## Exercise 07.4

We are now ready for the actual simulation of the system. The request of the exercise was of a simulation of $5\times10^5$ steps, but we will do only $2\times10^5$ steps due to the amount of time it takes. These steps are divided into $100$ blocks of $2000$ steps each; this should be enough to ensure a good result because of the considerations made before about the block size.

```python
import scipy.constants as sc
kb=sc.Boltzmann

sigma=0.34e-09
epsilon=120*kb
presunit=epsilon/sigma**3
```

```python
nblocks, epotS, err_epotS = np.loadtxt("data/solidoutput.epot.0", usecols=(0,2,3), unpack=True)
nblocks, presS, err_presS = np.loadtxt("data/solidoutput.pres.0", usecols=(0,2,3), unpack=True)
nblocks, epotL, err_epotL = np.loadtxt("data/liquidoutput.epot.0", usecols=(0,2,3), unpack=True)
nblocks, presL, err_presL = np.loadtxt("data/liquidoutput.pres.0", usecols=(0,2,3), unpack=True)
nblocks, epotG, err_epotG = np.loadtxt("data/gasoutput.epot.0", usecols=(0,2,3), unpack=True)
nblocks, presG, err_presG = np.loadtxt("data/gasoutput.pres.0", usecols=(0,2,3), unpack=True)

fig=plt.figure(figsize=(15,20))
plt.subplot(321)
plt.errorbar(nblocks, epotS*epsilon, yerr=err_epotS*epsilon)
plt.title("Solid potential energy")
plt.ylabel("Energy [J]")
plt.xlabel("# of blocks")
plt.grid(True)

plt.subplot(322)
plt.errorbar(nblocks, presS*presunit, yerr=err_presS*presunit)
plt.title("Solid pressure")
plt.ylabel("Pressure [Pa]")
plt.xlabel("# of blocks")
plt.grid(True)

plt.subplot(323)
plt.errorbar(nblocks, epotL*epsilon, yerr=err_epotL*epsilon)
plt.title("Liquid potential energy")
plt.ylabel("Energy [J]")
plt.xlabel("# of blocks")
plt.grid(True)

plt.subplot(324)
plt.errorbar(nblocks, presL*presunit, yerr=err_presL*presunit)
plt.title("Liquid pressure")
plt.ylabel("Pressure [Pa]")
plt.xlabel("# of blocks")
plt.grid(True)

plt.subplot(325)
plt.errorbar(nblocks, epotG*epsilon, yerr=err_epotG*epsilon)
plt.title("Gas potential energy")
plt.ylabel("Energy [J]")
plt.xlabel("# of blocks")
plt.grid(True)

plt.subplot(326)
plt.errorbar(nblocks, presG*presunit, yerr=err_presG*presunit)
plt.title("Gas pressure")
plt.ylabel("Pressure [Pa]")
plt.xlabel("# of blocks")
plt.grid(True)

plt.show()
```

For all the phases the code is able to give a measure of energy and pressure that stabilize around a definite value after about a bit more than half of the blocks with a reasonable statistical uncertainty.

During the simulation we have also computed average values for the function $g(r)$. In the next graphs we plot such function and compare it with the one obtained running the Molecular Dynamics NVE code in similar thermodynamic conditions.

```python
r, gofrS, err_gofrS = np.loadtxt("data/solidoutput.gave.0", usecols=(1,2,3), unpack=True)
gofr_NVES, err_gofr_NVES = np.loadtxt("data/solidoutput_g.out", usecols=(2,3), unpack=True)
r, gofrL, err_gofrL = np.loadtxt("data/liquidoutput.gave.0", usecols=(1,2,3), unpack=True)
gofr_NVEL, err_gofr_NVEL = np.loadtxt("data/liquidoutput_g.out", usecols=(2,3), unpack=True)
r, gofrG, err_gofrG = np.loadtxt("data/gasoutput.gave.0", usecols=(1,2,3), unpack=True)
gofr_NVEG, err_gofr_NVEG = np.loadtxt("data/gasoutput_g.out", usecols=(2,3), unpack=True)
r*=sigma

plt.figure(figsize = (15, 20))

plt.subplot(311)
plt.errorbar(r, gofrS, err_gofrS, label="NVT")
plt.errorbar(r, gofr_NVES, err_gofr_NVES, label="NVE")
plt.title("Solid g(r)")
plt.xlabel("r [m]")
plt.ylabel("$g(r)$")
plt.grid(True)
plt.legend()

plt.subplot(312)
plt.errorbar(r, gofrL, err_gofrL, label="NVT")
plt.errorbar(r, gofr_NVEL, err_gofr_NVEL, label="NVE")
plt.title("Liquid g(r)")
plt.xlabel("r [m]")
plt.ylabel("$g(r)$")
plt.grid(True)
plt.legend()

plt.subplot(313)
plt.errorbar(r, gofrG, err_gofrG, label="NVT")
plt.errorbar(r, gofr_NVEG, err_gofr_NVEG, label="NVE")
plt.title("Gas g(r)")
plt.xlabel("r [m]")
plt.ylabel("$g(r)$")
plt.grid(True)
plt.legend()

plt.show()
```

All the radial distributions show that the core of the potential is repulsive, since the probability of finding two particles closer than a certain value (that clearly depends on the phase we are considering) of $r$ is zero. We can observe that in the solid phase there are definite local maxima and minima, which are indicators of a cristalline structure. The liquid phase also shows a bit of this behavior, which is instead totally lost in the gas phase. From these plots the NVT and NVE codes seem to be in agreement except on the maximum of the gas phase.

We can investigate this further by plotting $g(r)_{NVT}-g(r)_{NVE}$:

```python
plt.figure(figsize = (15, 20))

plt.subplot(311)
plt.errorbar(r, gofrS-gofr_NVES, np.sqrt(err_gofrS**2+err_gofr_NVES**2))
plt.title("Solid $g(r)_{NVT}-g(r)_{NVE}$")
plt.xlabel("r [m]")
plt.ylabel("$g(r)_{NVT}-g(r)_{NVE}$")
plt.grid(True)

plt.subplot(312)
plt.errorbar(r, gofrL-gofr_NVEL, np.sqrt(err_gofrL**2+err_gofr_NVEL**2))
plt.title("Liquid $g(r)_{NVT}-g(r)_{NVE}$")
plt.xlabel("r [m]")
plt.ylabel("$g(r)_{NVT}-g(r)_{NVE}$")
plt.grid(True)

plt.subplot(313)
plt.errorbar(r, gofrG-gofr_NVEG, np.sqrt(err_gofrG**2+err_gofr_NVEG**2))
plt.title("Gas $g(r)_{NVT}-g(r)_{NVE}$")
plt.xlabel("r [m]")
plt.ylabel("$g(r)_{NVT}-g(r)_{NVE}$")
plt.grid(True)

plt.show()
```

From these plots we can see that, ignoring the maximum point, the gas phase is actually the one on which the two codes are more in agreement. For the other two phases the agreement is still good and interestingly it is worse for the values of $r$ in which the function $g(r)$ has maxima and minima points.
